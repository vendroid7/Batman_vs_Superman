```python
track_keyword = ['batman', 'superman']

def predict_sentiment(x):
    x = [x]
    x=tfidf.transform(x)
    sent = clf.predict(x.toarray())
    return sent



b=0
s=0



class MyStreamListener(tweepy.StreamListener):

    def on_status(self, status):
        print(status.text)
        
    def on_data(self,data):
        raw=json.loads(data)
#         print(raw['text'])
        
        try:
            x = str(raw['text']).lower()
#             x=ven.cont_exp(x)
#             x=ven.remove_html_tags(x)
#             x=ven.remove_emails(x)
#             x=ven.remove_rt(x)
#             x=ven.remove_special_chars(x)
#             x=ven.remove_urls(x)

            
            global b
            global s
            
            if 'batman' in x and 'superman' not in x:
                sent = predict_sentiment(x)
                b = b + sent[0]
#                 
                
            elif 'superman' in x and 'batman' not in x:
                sent = predict_sentiment(x)
                s = s + sent[0]
#                 
                
            else:
                pass
            
            print('Batman: ', b, 'Superman', s)
            
           
            
            
        except:
            pass
        
        
        
        
    def on_error(self, status_code):
        if status_code == 420:
           
            return False

        

myStreamListener = MyStreamListener()
myStream = tweepy.Stream(auth = api.auth, listener=myStreamListener)

myStream.filter(track=track_keyword)
```
